#### Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. Name, IMDB rating, Year of release) and make data frame.

```python
page_ind=requests.get("https://www.imdb.com/india/top-rated-indian-movies/")
```

```python
page_ind
```

```python
soup=BeautifulSoup(page_ind.text,'html.parser')
```

```python
soup
```

```python
ind_mov=soup.find('tbody',{'class':'lister-list'}).findAll('tr')
```

```python
ind_mov
```

```python
indian_movie=[]
for html in range(len(ind_mov)):
    if(html<=100):
        
        Name = ind_mov[html].find('td',{'class':'titleColumn'}).find('a').get_text(),
        Year_of_Release=ind_mov[html].find('td',{'class':'titleColumn'}).find('span').get_text(),
        Rating =ind_mov[html].find('td',{'class':'ratingColumn imdbRating'}).find('strong').get_text()
        data={
           'Name' :Name,
           'Year_of_Release':Year_of_Release,
           'Rating':Rating 
        }
        indian_movie.append(data)
```

```python
indian_movie=pd.DataFrame(indian_movie)
```

```python
def Internshala(url):
    p9 = requests.get(url)
    #p9.content
    print(p9.status_code)
    
    s9 = BeautifulSoup(p9.content, "html.parser")
    #print(s9.prettify())
    
    title = []
    company = []
    ctc = []
    apply_date = []
    
    temp = []
    
    get_titles = s9.find_all('div',class_ = "heading_4_5 profile")   
    get_companies = s9.find_all('a', class_ = 'link_display_like_text')
    get_ctcadate = s9.find_all('div',class_ = "item_body", id = False)
    
    for cdate in get_ctcadate:
        temp.append(cdate)
        
    for t, c in zip(get_titles, get_companies):
        title.append(t.text.strip())
        company.append(c.text.strip())
    
    for i in temp:
        ctc.append(i) if 'LPA' in i else apply_date.append(i)
        
    Internshala = pd.DataFrame({
         'Title': title,
          'Company': company,
#          'Apply date':temp,
         })
    print("Shape of dataset is {} rows and {} columns".format(*Internshala.shape))
    print(Internshala.head())
    #print(temp)
```

```python
Internshala("https://internshala.com/fresher-jobs")
```
