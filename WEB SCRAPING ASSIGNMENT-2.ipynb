2. In the above question, now scrape the following details of each product listed in first 3 pages 
of your search results and save it in a dataframe and csv. In case if any product has less than 3 
pages in search results then scrape all the products available under that product vertical. 
Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", 
"Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and 
“Product URL”. In case, if any of the details are missing for any of the product then replace it 
by “-“.


```python
urls=[i.get_attribute("href")for i in driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-4']//a")]
title = []
Brand =[]
prices = []
url = []
description = []
stocks = []
ratings = []
exp_del = []
for i in urls[0:100]:
    try:
        driver.get(i)
        time.sleep(2)
        raw_title = driver.find_element_by_xpath("//span[@class='a-size-large product-title-word-break']").text
        price = driver.find_element_by_xpath("//span[@class='a-size-medium a-color-price priceBlockDealPriceString']").text
        About_item = driver.find_element_by_xpath("//ul[@class='a-unordered-list a-vertical a-spacing-mini']").text
        stock = driver.find_element_by_xpath("//span[@class='a-size-medium a-color-success']").text
        rating = driver.find_element_by_xpath('//span[@id="acrPopover"]')
        expected_del = driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']//b")
        
    
        title.append(raw_title)
        Brand.append(raw_title.split()[0])
        prices.append(price)
        description.append(About_item)
        stocks.append(stock)
        url.append(i)
        ratings.append(rating.get_attribute("title"))
        exp_del.append(expected_del.text)
        
    except NoSuchElementException :
        title.append("---")
        Brand.append("---")
        prices.append("---")
        description.append("---")
        stocks.append("---")
        url.append("---")
        ratings.append("---")
        exp_del.append("---")
       
       
```

```python
pd.DataFrame({
    'Url' : url,
    'Brand': Brand,
    'products' :title,
    'price':prices,
    'ratings':ratings,
    'description':description,
    'stocks':stocks,
    'exp_del':exp_del
   
    
})
```

```python
BrandName = []
productName = []
price = []
color = []
storages=[]
display =[]
pcamera = [] 
scamera = []
resolution = []
battery = []
Name = driver.find_elements_by_xpath("//div[@class='_4rR01T']")
for i in Name:
    if i.text == '':
        BrandName.append("-")
        productName.append("-")
        color.append("-")
    else:
        BrandName.append(i.text.split()[0])
        productName.append(i.text)
        color.append(i.text.split()[4])
prices = driver.find_elements_by_xpath("//div[@class='_30jeq3 _1_WHN1']")
for i in prices:
    if i.text =='':
        price.append('-')
    else:
        price.append(i.text)
storage = driver.find_elements_by_xpath("//ul[@class='_1xgFaf']//li[1]")

for i in storage:
    if i.text == '':
        storages.append("-")
    else:
        storages.append(i.text.split()[0:2])

displays = driver.find_elements_by_xpath("//ul[@class='_1xgFaf']//li[2]")

for i in displays:
    if i.text == '':
        display.append("-")
        resolution.append("-")
    else:
        display.append(i.text.split()[0:4])
        resolution.append(i.text.split()[5:6])
cameras = driver.find_elements_by_xpath("//ul[@class='_1xgFaf']//li[3]")
for i in cameras:
    if i.text == '':
        pcamera.append("-")
        scamera.append("-")
    else:
        pcamera.append(i.text.split()[0:3])
        scamera.append(i.text.split()[4:6])
batterys = driver.find_elements_by_xpath("//ul[@class='_1xgFaf']//li[4]")
for i in batterys:
    if i.text == '':
        battery.append("-")
    else:
        battery.append(i.text)
processor = []
pros = driver.find_elements_by_xpath("//ul[@class='_1xgFaf']//li[5]")
for i in pros:
    if i.text == '':
        processor.append("-")
    else:
        processor.append(i.text)

        
links = driver.find_elements_by_xpath("//a[@class='_1fQZEK']")
link = []
for i in links:
    if i == '':
        link.append("-")
    else:
        link.append(i.get_attribute('href'))
        

pd.DataFrame({
    'url': link,
    'barnd':BrandName,
    'price':price,
    'color':color,
    'storages':storages,
    'display':display,
    'pcamera':pcamera,
    'scamera':scamera,
    'resolution':resolution,
    'battery':battery,
    'processor':processor
})
```
