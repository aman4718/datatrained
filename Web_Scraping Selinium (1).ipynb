```python
#### scrap full Job Description
get_link = driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')
get_link
```

```python
link = []
for i in get_link:
    link.append(i.get_attribute('href'))
link
```

```python
job_title = []
for i in link:
    driver.get(i)
    title = driver.find_elements_by_xpath("//h1[@class='jd-header-title']")
    for j in title:
        job_title.append(j.text)
job_title
```

```python
company_name = []
for i in link:
    driver.get(i)
    title = driver.find_elements_by_xpath("//a[@class='pad-rt-8']")
    for j in title:
        if j.text == "":
            continue
        else:
            
            company_name.append(j.text)
company_name
```

```python
job_desc = []
for i in link:
    title = driver.find_elements_by_xpath("//div[@class='clearboth description']/p")
    for j in title:
        job_desc.append(j.text)
job_desc
```

```python
job_desc = pd.DataFrame(job_desc).iloc[0:13]
job_desc
```

```python
job_details =pd.DataFrame({})
job_details['Title'] = job_title
job_details['Company']=company_name
job_details['Desc']=job_desc
job_details
```
